```python
import os

data_dir_source = '/home/wenjian/wiki_crop_original/'
target_root = '/home/wenjian/data/'
```

```python
def parse_age(image_name):
    '''
    example: 23300_1962-06-19_2011.jpg
    the calculation: age=2011-1962=49
    return 49
    '''
    tokens1 = image_name.split('.')
    tokens2 = tokens1[0].split('_')
    tokens3 = tokens2[1].split('-')
    dob = int(tokens3[0])
    photo = int(tokens2[2])
    return photo-dob
```

```python
def create_directories_according_to_age(target_dir, max_age=100):
    for i in range(max_age+1):
        try:
            os.mkdir(target_dir + str(i))
        except FileExistsError:
            pass
```

```python
from shutil import copyfile

def redistribute_image_according_to_age(source_dir, target_root, val_ratio=0.05, test_ratio=0.05):
    
    data_dir_train = target_root + 'wiki_crop_train/'
    data_dir_val = target_root + 'wiki_crop_val/'
    data_dir_test = target_root + 'wiki_crop_test/'
    try:
        os.mkdir(data_dir_train)
        os.mkdir(data_dir_val)
        os.mkdir(data_dir_test)
    except FileExistsError:
        pass
    
    create_directories_according_to_age(data_dir_train)
    create_directories_according_to_age(data_dir_val)
    create_directories_according_to_age(data_dir_test)
    
    print('Directories have been prepared.')
    
    directories = [x for x in os.listdir(source_dir) if os.path.isdir(source_dir+x)] # noted that the directories is not ordered
    count = 0
    train_val_border = int(100*(1-val_ratio-test_ratio)) #The border itself is included as val set
    val_test_border = int(100*(1-test_ratio)) #The border itself is included as test set
    for directory in directories:
        if int(directory) < train_val_border:
            target_dir = data_dir_train
        elif train_val_border <= int(directory) < val_test_border:
            target_dir = data_dir_val
        else:
            target_dir = data_dir_test
        for filename in os.listdir(source_dir + directory):
            try:
                age = parse_age(filename)
                if age < 0: # it seems that some image have negative age...
                    continue
                if age > 100: # it seems that some age are incorrectly large, like 269
                    continue
                copyfile(source_dir+directory+'/'+filename, target_dir+str(age)+'/'+filename)
            except IndexError :
                print('Failed to copy :' +directory+'/'+filename)
        if int(directory) < train_val_border:
            print(str(count), ' directory ' + directory + ' completed to train ')
        elif train_val_border <= int(directory) < val_test_border:
            print(str(count), ' directory ' + directory + ' completed to val ')
        else:
            print(str(count), ' directory ' + directory + ' completed to test ')
        
        count += 1
```

```python
redistribute_image_according_to_age(data_dir_source, target_root)
```
