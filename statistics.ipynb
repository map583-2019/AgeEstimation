```python
def get_distributions(path):

    distributions = {}

    for subdir in ['wiki_crop_train', 'wiki_crop_val', 'wiki_crop_test']:
        distribution = []
        for i in range(101):
            sub_path = path+subdir+'/'
            files = os.listdir(sub_path+str(i))
            distribution.append(len(files))
        distribution = np.array(distribution)
        distributions[subdir] = distribution
    return distributions
```

```python
dis_data4 = get_distributions('./data4/')
```

```python
def draw_distributions(distributions, title):
    plt.plot(distributions['wiki_crop_train'], label='wiki_crop_train')
    plt.plot(distributions['wiki_crop_val'], label='wiki_crop_val')
    plt.plot(distributions['wiki_crop_test'], label='wiki_crop_test')
    plt.title(title)
    plt.xlabel('age')
    plt.ylabel('image number')
    plt.legend()
```

```python
draw_distributions(dis_data4, 'data4')
```

```python
def total_number(distributions):
    return np.sum(distributions['wiki_crop_train'] + distributions['wiki_crop_val'] + distributions['wiki_crop_test'])
```
